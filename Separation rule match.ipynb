```python
import pandas as pd
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
import numpy as np
```

```python
HMP = pd.read_csv('for_pepe.csv')
```

```python
HMP1 = HMP[HMP['section'] == 1]
HMP2 = HMP[HMP['section'] == 2]
```

```python
first_FOMC1 = HMP1.groupby('meeting').first()['text']
last_FOMC1 = HMP1.groupby('meeting').last()['text']
first_FOMC2 = HMP2.groupby('meeting').first()['text']
last_FOMC2 = HMP2.groupby('meeting').last()['text']

second_FOMC2 = HMP2.groupby('meeting').nth(2)['text']
third_FOMC2 = HMP2.groupby('meeting').nth(3)['text']
second_last_FOMC2 = HMP2.groupby('meeting').nth(-2)['text']
third_last_FOMC2 = HMP2.groupby('meeting').nth(-3)['text']

second_FOMC1 = HMP1.groupby('meeting').nth(2)['text']
third_FOMC1 = HMP1.groupby('meeting').nth(3)['text']
second_last_FOMC1 = HMP1.groupby('meeting').nth(-2)['text']
third_last_FOMC1 = HMP1.groupby('meeting').nth(-3)['text']
```

```python
separation = pd.DataFrame(index = first_FOMC1.index, columns = ['FOMC1_1','FOMC1_2','FOMC1_3',
                                                                'FOMC1_-1','FOMC1_-2','FOMC1_-3',
                                                                'FOMC2_1','FOMC2_2','FOMC2_3',
                                                                'FOMC2_-1','FOMC2_-2','FOMC2_-3'])
```

```python
our_token =pd.read_excel('FOMC_token_all.xlsx')
our_token['Date'] = our_token['Date'].apply(lambda x: str(x)[:6])
```

```python
for meeting in separation.index:
    separation['FOMC1_1'][meeting] = get_index(first_FOMC1[meeting], meeting, our_token)
    separation['FOMC1_-1'][meeting] = get_index(last_FOMC1[meeting], meeting, our_token)
    separation['FOMC2_1'][meeting] = get_index(first_FOMC2[meeting], meeting, our_token)
    separation['FOMC2_-1'][meeting] = get_index(last_FOMC2[meeting], meeting, our_token)

    separation['FOMC1_2'][meeting] = get_index(second_FOMC1[meeting], meeting, our_token) -1
    separation['FOMC1_-2'][meeting] = get_index(second_last_FOMC1[meeting], meeting, our_token) +1
    separation['FOMC2_2'][meeting] = get_index(second_FOMC2[meeting], meeting, our_token) -1
    separation['FOMC2_-2'][meeting] = get_index(second_last_FOMC2[meeting], meeting, our_token) +1
    
    
    separation['FOMC1_3'][meeting] = get_index(third_FOMC1[meeting], meeting, our_token) -2
    separation['FOMC1_-3'][meeting] = get_index(third_last_FOMC1[meeting], meeting, our_token) +2
    separation['FOMC2_3'][meeting] = get_index(third_FOMC2[meeting], meeting, our_token) - 2
    separation['FOMC2_-3'][meeting] = get_index(third_last_FOMC2[meeting], meeting, our_token) +2
```

```python
separation.to_excel('Separation_new.xlsx')
```

```python
def get_index(x,meeting, our_token):
    
    meeting_token = our_token[our_token['Date'] == str(meeting)]['content'].values
    match = process.extractOne(x, meeting_token)
    print('meeting '+str(meeting))
    return np.where(meeting_token == match[0])[0][0]
```
